```python
import matplotlib.pyplot as plt
import pandas as pd
data=pd.read_csv('world_population_data.csv')
```

```python
year=['1980', '2000', '2010','2021', '2022', '2030', '2050']
pop=data[['1980', '2000', '2010','2021', '2022', '2030', '2050']]
```

```python
plt.plot(pop.sum(),linewidth=2)
plt.title('population growth')
plt.xlabel('year')
plt.ylabel('Bilions')
plt.grid()
plt.show()
```

```python
sorted=data.sort_values(by=year,ascending=False)
```

```python
plt.plot(sorted['country_code'].head(10),sorted['2050'].head(10),linewidth=2)
plt.title('top 10 countries population by 2050')
plt.xlabel('countries')
plt.ylabel('Bilions X 10')
plt.grid()
plt.figure().subplots_adjust(bottom=0.800)
plt.show()
```
